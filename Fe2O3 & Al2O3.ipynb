```python
# import packages
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
# import the data 
df = pd.read_excel('new.xlsx')
#showing the data
print (" Analyse the variation of Al2O3 & Fe2O3 with Depth \n so this the data provided to work on ")
print(df)
## refering the data

iron = df['Fe203 %']
Alu = df['Al203 %']
depth = df['Depth']

## calculating the mean and the median for iron

mean_iron = np.mean(iron)
median_iron = np.median (iron)
print(f'\nThe mean of Fe2O3 = {mean_iron} ' '\n' f'The median of Fe2O3 = {mean_iron}')

## creating classes 

num_classes = 7
data_range = iron.max() - iron.min()
class_width = data_range / num_classes
bins_lower = [iron.min() + i * class_width for i in range(num_classes + 1)]
bins_upper = [iron.max() + i * class_width for i in range(num_classes + 1)]
bins = [round(bins_lower[x],0)+round(bins_upper[x],0) for x in range (len (bins_lower))]  
    
classes = [val/2 for val in bins] ##midpoint of classes to represent on the graph

## calculating the mean and the median for Alumunium

mean_Alu = np.mean(Alu)
median_Alu = np.median (Alu)
print('\n'f'The mean of Al2O3 = {mean_Alu} ' '\n' f'The median of Al2O3 = {median_Alu}')


## frequency table
frequency_table , _ = np.histogram(iron , bins = 7)
print("\nThe Frequecy table :" , frequency_table , _ )

#cummulative frequency
cumm = np.cumsum(frequency_table)
print(f"The cummulative frequency : {cumm}")


## histogram 
plt.figure(figsize=(10, 6))
plt.bar(range(num_classes), cumm, width=1, edgecolor='black')
plt.xlabel('Classes MidPoints')
plt.ylabel('Cumulative Frequency')
plt.title('Histogram of Cumulative Frequency for Iron oxide')
plt.xticks(range(num_classes), classes[:-1])
plt.grid(True)
plt.show()

# Fe2O3 & Al2O3 against depth
plt.figure(figsize=(10, 6))
plt.scatter(depth  , iron, label =  'Fe2O3' )
plt.scatter (depth , Alu, label = 'Al2O3')
plt.xlabel('Depth (m)')
plt.ylabel('Concentraion %')
plt.legend()
plt.title('Fe2O3 % & Al2O3 % against depth')
```
